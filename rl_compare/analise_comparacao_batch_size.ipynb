```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme(style="darkgrid")

attr = ['time/total_timesteps','time/time_elapsed','rollout/ep_rew_mean']

df32 = pd.read_csv('results/compare_batch_size_32/progress.csv')
df32 = df32[attr]
df32['batch_size'] = 32

df64 = pd.read_csv('results/compare_batch_size_64/progress.csv')
df64 = df64[attr]
df64['batch_size'] = 64

df128 = pd.read_csv('results/compare_batch_size_128/progress.csv')
df128 = df128[attr]
df128['batch_size'] = 128

df256 = pd.read_csv('results/compare_batch_size_256/progress.csv')
df256 = df256[attr]
df256['batch_size'] = 256

df512 = pd.read_csv('results/compare_batch_size_512/progress.csv')
df512 = df512[attr]
df512['batch_size'] = 512

df = pd.concat([df32,df64,df128,df256,df512])
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="time/total_timesteps", y="rollout/ep_rew_mean", hue='batch_size', data=df)
plt.xlabel('Timesteps')
plt.ylabel('Recompensa')
plt.title('Comparando diferentes tamanhos de batch size no algoritmo DQN')
plt.legend(title='Tamanho de batch size')
plt.show()
```
