```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="time/episodes", y="rollout/ep_rew_mean", hue='replay_buffer', data=df)
plt.xlabel('Episódios')
plt.ylabel('Recompensa')
plt.title('Comparando diferentes tamanhos de replay buffer no algoritmo DQN')
plt.legend(title='Tamanho do replay buffer')
plt.show()
```
